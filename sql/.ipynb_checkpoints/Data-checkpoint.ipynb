```python
def genName():
	return ' '.join([choice(ho), choice(dem), choice(ten)])

def genCMND():
	return ''.join(choices(digits, k=12))

def genDate():
	return "{}-{}-{}".format(randint(1990, 2005), randint(1,12),randint(1,28))

def genSDT():
	tmp = choice(sdt)
	return ''.join(list(tmp) + choices(digits, k=(10 - len(tmp))))

def genDiaChi():
	return "{} Đường {}, Phường {}, Quận {}, TP.Hồ Chí Minh".format(randint(1, 1000), choice(duong), choice(phuong), choice(quan))
```
